```python
# prompt: buatkan aku codingan neural network yang pastinya bisa difahami alur codenya oleh orang yang awam pada neural network,bisakah kamu buat seperti perupamaan di kehidupan sehari2 agar lebih mudah difahami oleh orang yang dongo matematika

# Bayangkan kita punya tukang kopi yang mau belajar bikin kopi enak.
# Dia punya beberapa "pengalaman" (data training) tentang biji kopi, air, dan rasa.

# Input: Biji kopi (misal: robusta, arabica), jumlah air, suhu air
# Output: Rasa kopi (misal: pahit, asam, manis)

import numpy as np

# Ini "otak" tukang kopi, di dalamnya ada "sel-sel saraf" (neuron)
# yang akan diubah-ubah selama dia belajar.
weights = np.array([0.5, 0.2, 0.3])  # Awalnya random, kayak tukang kopi baru belajar
bias = 0.1  # Kayak "suka tidak suka" awal tukang kopi terhadap kopi

# Fungsi aktivasi, kayak cara tukang kopi menilai rasa
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

# Fungsi untuk menebak rasa kopi
def predict(input_data):
  # Tukang kopi ngitung "skor rasa" dari inputnya
  weighted_sum = np.dot(input_data, weights) + bias
  # Skor rasa itu diubah jadi penilaian (0-1) lewat fungsi sigmoid
  output = sigmoid(weighted_sum)
  return output

# Ini data training, beberapa contoh biji kopi, air, dan rasa
training_data = np.array([
  [1, 0.5, 80], # Robusta, air sedang, suhu tinggi --> Rasa pahit (1)
  [0, 0.3, 90], # Arabica, air sedikit, suhu tinggi --> Rasa asam (0)
])

target_values = np.array([1, 0]) # target nilai rasa, pahit (1) asam (0)

# Proses pembelajaran, kayak tukang kopi coba-coba bikin kopi
learning_rate = 0.1  # Seberapa cepat dia belajar dari kesalahannya
epochs = 1000 # Jumlah percobaan

for epoch in range(epochs):
  for i in range(len(training_data)):
    input_data = training_data[i]
    target_value = target_values[i]

    # Tukang kopi menebak rasa dari inputnya
    predicted_value = predict(input_data)

    # Tukang kopi membandingkan tebakannya dengan rasa sebenarnya
    error = target_value - predicted_value

    # Tukang kopi menyesuaikan "otaknya" (weights) berdasarkan kesalahan
    weights = weights + learning_rate * error * predicted_value * (1 - predicted_value) * input_data
    bias = bias + learning_rate * error * predicted_value * (1 - predicted_value)


    #________________-Sesudah Belajar dan eksperimen_______________

new_input = np.array([0.9, 0.9, 99]) # Contoh input baru
predicted_taste = predict(new_input)

print("Prediksi rasa kopi:", predicted_taste)

# Nilai prediksi mendekati 1 artinya rasa pahit, mendekati 0 artinya rasa asam
```
